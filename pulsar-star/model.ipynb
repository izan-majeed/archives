## <font color='#3caea3'> Train Test Split </font> 

```python
from sklearn.model_selection import train_test_split
```

```python
X = my_data.drop('target_class', axis=1)
y = my_data['target_class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
y_train.value_counts()
```

## <font color='#ff6f69'> Standardization </font> 

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler() 
s.fit(X_train)
```

```python
X_train_std = s.transform(X_train)
X_test_std = s.transform(X_test)
```

## <font color='#3caea3'> Models </font> 

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(C=10)
log_model.fit(X_train_std, y_train)
log_pred = log_model.predict(X_test_std)
```

```python
from sklearn.svm import SVC

svc_model = SVC(C=10, degree=2, gamma='auto')
svc_model.fit(X_train_std, y_train)
svc_pred = svc_model.predict(X_test_std)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_model = RandomForestClassifier()
forest_model.fit(X_train, y_train)
forest_pred = forest_model.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()
knn_model.fit(X_train_std, y_train)
knn_pred = knn_model.predict(X_test_std)
```

## <font color='#ff6f69'> Tuning </font> 

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_log = {
    'C':[0.1, 1.0, 10],
    'solver':['lbfgs', 'newton-cg'] }

param_grid_svc = {
    'C':[0.1, 1, 10],
    'degree':[2, 3, 4],
    'gamma':['scale', 'auto'], }

param_grid_forest = {
    'n_estimators':[50, 100, 200],
    'max_features':['auto', 2, 3, 4, 5],
    'bootstrap':[True, False], }
```

```python
grid_search_log = GridSearchCV(LogisticRegression(), param_grid_log, cv=3)
grid_search_svc = GridSearchCV(SVC(), param_grid_svc, cv=3)
grid_search_forest = GridSearchCV(RandomForestClassifier(), param_grid_forest, cv=3)
```

```python
grid_search_log.fit(X_train_std, y_train)
grid_search_svc.fit(X_train_std, y_train)
grid_search_forest.fit(X_train, y_train)
```

```python
grid_search_log.best_params_
```

```python
grid_search_svc.best_params_
```

```python
grid_search_forest.best_params_
```

```python
from sklearn.metrics import accuracy_score
```

```python
error_rate = []
for i in range(1, 60):
    cfc = KNeighborsClassifier(n_neighbors=i)
    cfc.fit(X_train_std, y_train)
    ith_pred = cfc.predict(X_test)
    ith_accuracy = accuracy_score(y_test, ith_pred)
    error_rate.append(1-ith_accuracy)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 60), error_rate, 'r--', marker='o')
plt.show()
```

## <font color='#3caea3'> Evaluation </font> 

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, log_pred))
print('\n')
print(classification_report(y_test, log_pred))
```

```python
print(confusion_matrix(y_test, svc_pred))
print('\n')
print(classification_report(y_test, svc_pred))
```

```python
print(confusion_matrix(y_test, forest_pred))
print('\n')
print(classification_report(y_test, forest_pred))
```

```python
print(confusion_matrix(y_test, knn_pred))
print('\n')
print(classification_report(y_test, knn_pred))
```
